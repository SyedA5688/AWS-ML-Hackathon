# Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
import joblib
import random
```

# Load in Data Files

```python
npTrainMatrix = np.load('./data/2018-01-01__2019-01-01__NConservatory_npWeekdayAllOrderedSensorsTimeRef.npy', allow_pickle=True)
npTestMatrix = joblib.load('./test_data/test_dataset_2019.numpy')
npTrainMatrix.shape, npTestMatrix.shape
```

```python
print("Sample of training (left) and testing (right) data - sensor values from 1 day of readings")
plt.figure(figsize = (15,4))

plt.subplot(1,2,1)
plt.plot(npTrainMatrix[0,:])

plt.subplot(1,2,2)
plt.plot(npTestMatrix[0])

plt.show()
```

# Insert Anomalies into Training Data for Isolation Forest Algorithm

```python
# Function takes in dataset and a percentage (0-1) of samples to insert anomalies into, inserts anomalies into first - samples

def insert_anomalies(npMatrix, percent):
    numAnomsToInsert = int(npTrainMatrix.shape[0] * percent)
    readingsPerSensor = 96
    randSensor = random.randint(0,15) * readingsPerSensor # Get index of where one of the first 16 sensors' readings start
    
    for i in range(numAnomsToInsert):
        for j in range(readingsPerSensor):
            # Change sensor readings by 2
            if (i % 2 == 0):
                npMatrix[i][j] += 2
            else:
                npMatrix[i][j] -= 2
    
    print(numAnomsToInsert, "anomalies inserted into dataset")
```

```python
# Isolation Forest algorithm takes contamination parameter, a percent of anomalies to expect in data. Testing data has 200/9898
# anomalies, 2%, or 0.02, so insert anomalies into 2% of training data

np.random.shuffle(npTrainMatrix) # Shuffle rows
insert_anomalies(npTrainMatrix, 0.02) # Insert anomalies into 2% of training samples
np.random.shuffle(npTrainMatrix) # Shuffle again to disperse anomalies
```

# Build & Train Isolation Forest Model

```python
rng = np.random.RandomState(42)
model = IsolationForest(max_samples='auto', random_state=rng, contamination=0.02)
model.fit(npTrainMatrix)
```

# Use Model to Make Predictions on Test Data

```python
y_pred_train = model.predict(npTrainMatrix)
y_pred_test = model.predict(npTestMatrix)
y_pred_train.shape, y_pred_test.shape
```

# Evaluate Model

```python
# Training set does not have any anomalies, count_anom should be 0
train_count_anom = list(y_pred_train).count(1)/y_pred_train.shape[0]
print("Accuracy: ", train_count_anom)
```

```python
# Test set has 200 anomalies injected into dataset
test_count_anom = list(y_pred_test).count(-1)
print("Accuracy of test evaluation:", test_count_anom, "/", 200, "anomalies detected")
```
